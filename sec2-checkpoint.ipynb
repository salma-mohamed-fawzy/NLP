```python
import nltk;
from nltk.corpus import gutenberg
def unusual_text(text):
    text_vocab=set(i.lower() for i in text if i.isalpha())
    english_vocab=set(i.lower() for i in nltk.corpus.words.words() if i.isalpha())
    
    unusual_text=english_vocab.difference(text_vocab)
    return sorted (unusual_text)
unusual_text(nltk.corpus.gutenberg.words('austen-sense.txt'))
```
